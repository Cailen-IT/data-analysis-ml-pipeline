# Data Analysis & Machine Learning Pipeline
This project demonstrates:
- Data cleaning and preprocessing
- Exploratory Data Analysis (EDA)
- Feature engineering (Revenue calculation)
- Data visualization and insights generation



```python
import pandas as pd

df = pd.read_csv("data/sales_data.csv")  # adjust if filename differs
df.head()
df.info()


# Convert Date to proper datetime format
df["Date"] = pd.to_datetime(df["Date"])

# Create a new column: Revenue (this is what makes it "real analysis")
df["Revenue"] = df["Units_Sold"] * df["Unit_Price"]

df.head()
```

```python
df.groupby("Category")["Revenue"].sum()
```

```python
import matplotlib.pyplot as plt

df.groupby("Category")["Revenue"].sum().plot(kind="bar")
plt.title("Revenue by Category")
plt.show()
```

#Key Insight

Revenue is significantly driven by product category differences, showing that certain categories contribute more to overall sales performance. This can guide business decision-making and resource allocation.
